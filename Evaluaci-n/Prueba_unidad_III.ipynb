# üõ§Ô∏è Ejercicio 1: Control de Tren (Valor - 6 pts)

### Descripci√≥n
Simula el sistema de control de un tren que debe responder a se√±ales en la v√≠a.  
Las se√±ales pueden ser **"verde"**, **"amarillo"** o **"rojo"**.  
El tren debe ajustar su velocidad base seg√∫n la se√±al actual y la distancia hasta la pr√≥xima estaci√≥n:

- **Se√±al verde**:  
  - Acelera a **100 km/h** si la distancia es mayor a 5 km.  
  - Mantiene **80 km/h** en caso contrario.  

- **Se√±al amarilla**:  
  - Reduce la velocidad a **50 km/h** si la distancia es mayor a 3 km.  
  - De lo contrario, reduce a **30 km/h**.  

- **Se√±al roja**:  
  - Frena hasta **0 km/h** inmediatamente.  

- **Obst√°culo en la v√≠a**:  
  - El tren debe frenar a **0 km/h**, independientemente de la se√±al.  

---

### ‚úÖ Salida esperada
El programa debe devolver la velocidad actual del tren en base a las condiciones dadas.  

**Ejemplo:**
- Entrada: `se√±al="amarillo", distancia=4, obst√°culo=False`  
- Salida: `Velocidad: 50 km/h`

```python

#Pregunte al ususario la distaancia
Distancia= int(input("Ingrese la distancia en kilometros: "))

# Pregunte al usuario si hay un obstaculo
Obstaculo= bool(input("¬øHay un obst√°culo en el camino? Si no hay no escriba nada: "))

#Categoriza la respuesta en forma booleana para vver si hay o no obstaculos
if Obstaculo == True:
    ObsStatus = "Hay obstaculos en el camino"
else:
    ObsStatus = "No hay obstaculos en el camino"

# Imprime dist y obstaculo para el usuario
print(f"La distancia es igual a {Distancia} km y {ObsStatus}")

#Verifica el estado del obstaculo y si es True detiene el programa luego de detener el tren
if Obstaculo == True:
    
    print("Detengase ahora mismo.")
    quit()
else:
    print("Puede continuar su camino.")
    

# Pregunta al usuario el estado del semaforo
print("Estado del semaforo: Amarillo [1], Rojo [2], Verde [3]")
SmfrClr = int(input("Ingrese el estado del semaforo: "))

# Calcula la velocidad segun el estado del semaforo y la distancia
if SmfrClr == 1:
    print("El sem√°foro est√° en amarillo.")
    if Distancia > 3:
        print("Reduzca la velocidad a 50 km/h.")
    else:
        print("Reduzca la velocidad a 30 km/h.")
        
elif SmfrClr == 2:
    print("El sem√°foro est√° en rojo.")
    print("Reduzca a 0 km/h inmediatamente.")
elif SmfrClr == 3:
    print("El sem√°foro est√° en verde.")
    if Distancia > 5:
        print("Acelere a 100 km/h.")
    else:
        print("Mantenga la velocidad a 80km/h")
else:
    print("Estado de sem√°foro no v√°lido.")
    
# Fin del programa
```

# üí∞ Ejercicio 3: C√°lculo de Impuestos (Valor - 7 pts)

### Descripci√≥n
Calcula el **impuesto sobre la renta** en base al **ingreso anual**, la **edad** y el **n√∫mero de dependientes**.  

Las reglas son:

- Ingresos **‚â§ $10,000**: no pagan impuestos.  
- Ingresos **entre $10,001 y $50,000**: pagan **10% sobre el exceso** de $10,000.  
- Ingresos **> $50,000**: pagan **20% sobre el exceso de $50,000** m√°s **$4,000**.  
- Mayores de **65 a√±os**: reducci√≥n del **5% en el impuesto total**.  
- Por cada **dependiente**, se reduce **$1,000 del ingreso gravable** antes de calcular el impuesto.  

---

### ‚úÖ Salida esperada
El programa debe calcular y mostrar el impuesto a pagar.  

**Ejemplo:**
- Entrada: `ingreso=60000, edad=70, dependientes=2`  
- Salida: `Impuesto: 5320`  

üìå *Explicaci√≥n:*  
- Ingreso gravable = 60000 - (1000 √ó 2) = **58000**  
- Impuesto base = 4000 + 20% de (58000 - 50000) = **5600**  
- Descuento por edad = 5% de 5600 = **280**  
- Impuesto final = 5600 - 280 = **5320**

```python
 
ingreso = int(input("¬øCu√°l es tu ingreso anual? "))
edad = int(input("¬øCu√°ntos a√±os tienes? "))
dependientes = int(input("¬øCu√°ntos dependientes tienes? "))


 
descuento_dependientes = dependientes * 1000
ingreso_gravable = ingreso - descuento_dependientes

 
if ingreso_gravable < 0:
    ingreso_gravable = 0

print(f"\nIngreso Gravable (despu√©s de dependientes): ${ingreso_gravable}")


impuesto_base = 0


if ingreso_gravable <= 10000:
    impuesto_base = 0 
    print("El ingreso gravable es menor o igual a $10,000. Impuesto base: $0")


elif ingreso_gravable <= 50000:
    #
    exceso = ingreso_gravable - 10000
    impuesto_base = exceso * 0.10 
    print(f"El ingreso gravable est√° entre $10,001 y $50,000.")
    print(f"Exceso sobre $10,000: ${exceso}. Impuesto base (10%): ${impuesto_base}")


else:
    
    impuesto_fijo = 4000
    exceso = ingreso_gravable - 50000
    impuesto_exceso = exceso * 0.20   
    impuesto_base = impuesto_fijo + impuesto_exceso
    print(f"El ingreso gravable es mayor a $50,000.")
    print(f"Impuesto fijo: ${impuesto_fijo}. Exceso sobre $50,000: ${exceso}.")
    print(f"Impuesto sobre exceso (20%): ${impuesto_exceso}. Impuesto base total: ${impuesto_base}")

 
impuesto_final = impuesto_base
descuento_edad = 0

if edad > 65:
    
    descuento_edad = impuesto_base * 0.05   
    impuesto_final = impuesto_base - descuento_edad
    print(f"\n¬°Felicidades! Tienes m√°s de 65 a√±os.")
    print(f"Descuento por edad (5% de ${impuesto_base}): ${descuento_edad}")

 
impuesto_a_pagar = round(impuesto_final)

print("\n" + ("-" * 30))
print(f"‚úÖ **Impuesto final a pagar: ${impuesto_a_pagar}**")
print("-" * 30)
```
